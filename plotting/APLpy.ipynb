# APLpy

This is basically a module that understands FITS files, understands astronomical coordinate systems, and adds this intelligence to Matplotlib.

Documentation is at https://aplpy.readthedocs.io/en/stable/index.html

Being able to zoom and pan is useful, so we use the notebook backend rather than inline.

```python
%matplotlib notebook
import aplpy
display(aplpy.__version__)
```

```python
from astropy.io import fits
from astropy.utils.data import download_file
```

Get some 2MASS data from astropy. This one is with the H filter (infrared). The data is cached in your home directory and only needs to be downloaded once.

```python
image_h = download_file('https://astropy.stsci.edu/data/galactic_center/gc_2mass_h.fits', cache=True )
image_h
```

```python
gc_h = aplpy.FITSFigure(image_h)
gc_h.show_grayscale()
gc_h.add_grid()
```

## What else?

APLpy will handle a variety of overlays, and will interface with [Montage](http://montage.ipac.caltech.edu/) to make this more sophisticated. However, it is at heart a plotting program that very successfully sets out to do one thing well, with the minimum of fuss.
