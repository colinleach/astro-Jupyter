## Pressure

For the atmosphere to exist it must be in _hydrostatic equilibrium_, with a pressure gradient balancing gravity. The general equation for this pressure gradient is $\displaystyle \frac{dP}{dh} = -g \rho $

There are no constants in this equation: gravitational acceleration $g$ is a function of distance from the planet's center, and hence of height. Density $\rho$ depends on pressure (and hence height), but also on temperature and chemical composition.

### Equation of state

For any gas, pressure, volume and temperature are not all independent quantities. In general there will be a function $f$ such that $f(P,V,T) = 0$.

#### Ideal gas

The simplest and most familiar case is the ideal gas law $P = n k_B T$, so $f(P,V,T) = P-k_B T = 0$. 

There are many equivalent ways to express this, though important distinctions can get lost behind confusingly similar and often ambiguous notations. If $N$ is the total number of particles (dimensionless) then $n = N/V$ is the number density ($m^{-3}$), $N_A \approx 6 \times 10^{23}$ is Avogadro's number and $n_0 = N/N_A$ is the mole number. 

Also, the universal gas constant is $R = N_A k_B$ and the specific gas constant is $R_g = R/M = k_B/\mu$ where $M$ is the molar mass (kg/mol) and $\mu$ is the molecular mass (aka molecular weight, in atomic mass units $\equiv$ Da).

Thus we can also say that 
$$P V = N k_B T = n R T = n_0 R_g T$$

More usefully for astronomers (who try to avoid volume terms):
$$P = \rho \left( \frac{R}{\mu} \right) = \rho R_g T$$
where $\rho$ is the density ($kg/m^3$).

It is also common to relate these terms to the specific heat capacity at constant pressure ($c_P$) and constant volume ($c_V$). Then the heat capacity ratio is 

$$ \gamma = \frac{c_P}{c_V} $$

and the specific gas constant is $R_g = c_P - c_V$. 

For an ideal monatomic gas $\gamma = 5/3 \approx 1.67$ and for a diatomic gas (such as Earth's lower atmosphere) $\gamma = 1.4$

The speed of sound is 
$$ c = \sqrt{\gamma \frac{P}{\rho}} = \sqrt{\gamma R_g T} $$

__What makes an ideal gas?__ Aside from the tautology that it is something that obeys the ideal gas law to good approximation, the assumptions are:
- non-relativistic motion
- particles interact only through collisions
- particles have no significant internal structure

So not too hot, not too dense, not too much chemical complexity. The atmospheres of Earth and Mars pretty much qualify, the interior of Jupiter really doesn't.

#### Polytropes

These are systems in which pressure and density take the form of an temperature-independent power law. This can be written various ways, but commonly:

$$ P = K \rho^{(n+1)/n} $$

where $n$ is the polytropic index (nothing to do with number density).

#### Degenerate gas

At high densities the particles of "gas" are so closely packed that quantum mechanical effects become more important than thermal motions. In particular, fermions (particles with half-integer spin such as electrons, protons and neutrons) obey the Pauli exclusion principle and cannot share the same quantum state. This is most obvioiusly relevant to stellar remnants: white dwarves supported by electron degeneracy pressure and neutron stars by neutron degeneracy. However it can also be a factor in gas giants including Jupiter.

A key point about degenerate states is that pressure is independent of temperature. For the non-relativistic case:
$$ P = \frac{(3 \pi^2)^{2/3}}{5} \frac{\hbar^2}{m} n^{5/3} $$ 
where $n$ is the number density.

If the gas is ultra-relativistic with momentum $p \gg mc$ (unlikely for planets):
$$ P = \frac{(3 \pi^2)^{1/3}}{4} \hbar c n^{4/3} $$

Why do planetary scientists care? One odd feature of electron degeneracy is that volume is ___inversely___ proportional to mass, as self-gravity wins out over internal pressure. Jupiter may be massive enough and dense enough that adding mass would shrink its radius. 

### The exponential atmosphere

This simplest model of atmospheric pressure makes two simplifying assumptions:
- gravitational acceleration is constant with height (reasonable for the shallow atmospheres of rocky planets)
- temperature is constant with height (a pretty gross approximation - see the next section)

Given this we can solve a simple first-order ODE to get 
$$ P(h) = P_0\ e^{-h/H} $$
where $P_0$ is the pressure at the surface and $H$ is the _scale height_, the vertical distance giving an $e$-fold drop in pressure.

Slightly more complex models allow $H$ to vary with height. Then the $\exp(-h/H)$ term becomes $\displaystyle \exp \left( -\int \frac{dh}{H(h)} \right)$

## Temperature

This is much less simple and predictable than pressure. On Earth we rely on frequent and detailed measurements to make sense of it, including the launch of about 1600 weather balloons per day (800 sites worldwide, 12-hour intervals). Good luck trying to make similar models for Venus or Jupiter.

### Controlling factors

Anyone who has taken an introductory astronomy course has probably done the standard Stefan-Boltzmann calculation: at equilibrium the solar energy falling on the planet must equal blackbody emission. A planet with albedo $a$ at distance $d$ from a star of radius $R_s$ and surface temperature $T_s$ should have a temperature of 
$$ T_p = T_s (1-a)^{1/4} \sqrt{\frac{R_s}{2d}} $$

If this was accurate, Venus should be habitable for humans and Earth too cold. There must be other things to consider.

#### Greenhouse effect

The simple calculation assumes both star and planet are (spherically symmetric) black body emitters and doesn't involve the wavelengths of light from each. As stars are much hotter than planets they emit at much shorter wavenegths (visible vs IR for our Solar System). Atmospheres have very different optical depths at different wavelengths, so a greenhouse effect occurs whenever the atmosphere is transparent to incoming solar radiation and relatively opaque to outgoing planetary radiation. Very common! 

Several typical components of atmospheres have strong IR absorption bands (water vapor, carbon dioxide, methane and others), which keeps us warm on Earth but also motivates astronomers to spend a lot of money on launching space telescopes. In a more extreme case, the surface of Venus is hot enough to melt lead (or whatever is holding the components onto your lander's electronic circuit boards).

Read in some published planetary data then calculate the expected temperature. Note that these are all the objects for which I could find temperature and albedo values; I am not claiming that Pluto is a "planet".

```python
planets = Table.read('../files/planets.fits')
planets.sort('a')

T_sun = 5778*u.K
planets['T_p'] = (T_sun * (1-planets['albedo'])**(1/4) * np.sqrt(R_sun/(2*planets['a']))).to(u.K)
planets['Planet', 'a', 'albedo', 'T_mean', 'T_p']
```

Next plot both measured and calculated temperatures for comparison. We can see that the calculations work well for Mercury, Mars and Pluto, all with little or no atmosphere (around 7 mb surface pressure on Mars, much less for the others).

Other planets are all warmer than the calculation suggest. Venus is ___by far___ the wildest deviation. 

```python
fig = plt.figure(figsize=(8, 8))
plt.plot(planets['a'], planets['T_mean'], 'bo', label='measured')
plt.plot(planets['a'], planets['T_p'], 'r^', label='calculated')
plt.plot((0.35, 40), (273, 273), 'k:', label='MPt of water ice')
plt.xscale('log')
plt.xlabel('Distance from Sun (AU)', fontsize=18)
plt.ylabel('Temp (K)', fontsize=18)
plt.legend();
```

#### Internal planetary processes

Rocky planets, and presumably rocky/metallic cores in gas giants, produce heat from nuclear fission. This doesn't seem to have a major direct impact on Earth's atmospheric temperature and composition, but by driving volcanism the resulting outgassing and aerosol formation may be signficant?

Gas giants can have atmospheric pressures into the Mbar range, a strange and unfamilar part of the hydrogen/helium phase diagrams. Saturn seems to release a lot of gravitational energy as He droplets sink towards the core, separating from an H<sub>2</sub>-rich upper atmosphere..

### Lapse rate

This is simply $\frac{dT}{dh}$, the temperature change per unit height change, with units such as K/km. Negative values mean high points are colder than low points, as near the Earth's surface (obvious when you drive to a mountain-top telescope). In practice the lapse rate can be a complicated function of height with multiple changes of sign: see the section on Earth's atmosphere.

### Adiabatic lapse rate and convection

As a parcel of (ideal) gas rises it expands (because of lower pressure) and cools (because it does work on its surroundings). At equilibrium, for an adiabiatic process (no energy added or removed) the lapse rate will be
$$ \frac{dT}{dh} = -\frac{g}{C_p} $$
where $g$ is the local gravitational acceleration and $C_p$ is the specific heat at constant pressure.

How the actual lapse rate differs from the adiabatic lapse rate determines atmospheric stability. If $dT/dh + g/C_p$ is positive the atmosphere is stable: a rising parcel of air finds itself denser than the surrounding air and falls back. If $dT/dh + g/C_p$ is negative the atmosphere is convectively unstable: small vertical displacements tend to grow.

This assumes no condensation, so more precisely we are talking about the _dry_ adiabatic lapse rate.

## Atmosphere loss

How atmospheres form in the first place is a large separate topic. Here we are just considering how volatiles are subsequently lost to space by escape or erosion.

The escape region at the top of the atmosphere is called the _exosphere_ and its lower limit is the _exobase_. In planets with thin atmospheres the exobase may be at the surface (see Sanchez-Lavega, table 2.2).

### Thermal escape

Molecules from the exophere will be lost into space if their velocity exceeds the escape velocity.

Escape velocity is simple, as we presumably know the mass of the planet and the (planet + atmosphere) radius:
$$ v_{esc} = \sqrt \frac{2GM}{r} $$

The `planets` table already has a `v_esc` column. Using the above formula to calculate it from `M` and `R_eq` (not shown here) gives good agreement except for Saturn, where `R_eq` unfortunately includes the rings.

```python
planets['Planet', 'v_esc']
```

Atomic/molecular velocity is a matter of statistical mechanics and a bit more complicated. The distribution of velocity $v$ for a particle of mass $m$ at temperature $T$ follows the Maxwell-Boltzmann velocity distribution function:

$$ p(v) = \left( \frac{m}{2 \pi k T} \right) e^{-mv^2/2kT} 4 \pi v^2 $$

That's hard to visualize, so several volatiles are plotted below. 

```python
from astropy.constants import k_B, m_p, G

def p_v(m, T, v):
    return ((m/(2*np.pi*k_B*T))**1.5 * np.exp(-m*v**2/(2*k_B*T)) * 4*np.pi*v**2).to(u.s/u.km)
```

```python
v = np.linspace(0, 2e4, 200)*u.m/u.s
v_km_s = v.to(u.km/u.s) # just for plotting

# approximate atom/molecule weights
m_H2 = 2*m_p
m_He = 4*m_p
m_N2 = 28*m_p

# escape velocities (km/s)
v_esc_earth = 11.2
v_esc_mars = 5.0

def plot_vdf(T):
    T *= u.K
    vdf_H2 = p_v(m_H2, T, v)
    vdf_He = p_v(m_He, T, v)
    vdf_N2 = p_v(m_N2, T, v)

    fig = plt.figure(figsize=(12, 8))
    plt.plot(v_km_s, vdf_H2, 'b-', label='H2')
    plt.plot(v_km_s, vdf_He, 'g-', label='He')
    plt.plot(v_km_s, vdf_N2, 'r-', label='N2')
    ymax = vdf_N2.max().value
    plt.plot((v_esc_earth, v_esc_earth), (0, ymax), ls='dotted', label='v_max Earth')
    plt.plot((v_esc_mars, v_esc_mars), (0, ymax), ls='dashdot', label='v_max Mars')
    plt.xlabel('velocity (km/s)', fontsize=18)
    plt.ylabel('Probability density (s/km)', fontsize=18)
    plt.title(f'Velocity distribution functions at {T}')
    plt.legend();
```

```python
style = {'description_width': 'initial'} # to avoid the labels getting truncated
interact(plot_vdf, 
             T = w.FloatSlider(description="Temperature (K)", style=style,
                                            layout=Layout(width='80%'),
                                            continuous_update=False, 
                                            min=50, max=1500, 
                                            value=300));
```

From plots like this it doesn't exactly seem obvious that Earth should have lost most of its hydrogen and Mars should have little atmosphere. There must be other important factors.

As escape is presumably slow (Myr or Gyr timescales), one parameter of interest is the area under the curve for $v_{esc} < v < \infty$. Carrol & Ostlie take this a stage further by defining am _atmospheric escape parameter_ $\nu$ with units of velocity:

$$ \nu = \frac{1}{8} \left( \frac{m}{2 \pi k T} \right)^{1/2} \left( v_{esc}^2 + \frac{2kT}{m} \right) e^{-mv_{esc}^2/2kT} $$

Then the number escaping per unit time is $ \dot{N}(z) = [4 \pi R^2 n(z)]\ \nu $, where $n(z)$ is the particle number density.

### Non-thermal escape

A variety of processes can give individual molecules or ions a kinetic energy significantly greater than the thermal distribution of the surrounding bulk medium. The relative significance of these can depend on the (variable) activity of the Sun and the presence or absence of a planetary magnetic field (e.g. Earth vs Mars).

__Photochemical escape:__ driven by UV irradiation, photodissociation can give fragments with enough kinetic energy to escape. Important for loss of C/O/N from Mars.

__Charge exchange__: a fast ion collides with a neutral species, losing charge to become a fast neutral atom that can escape. Important for loss of H from Earth.

__Ion pickup:__ acceleration of an atmospheric ion by an electric field from the magnetized solar wind.

__Sputtering:__ high energy particles (from cosmic rays, or in the magnetosphere of giant planets) impact the atmosphere or surface, causing a cascade of collisions that can result in some atoms having the energy to escape. Probably important on early Mars after it lost its magnetic field, and on the satellites of Jupiter and Saturn.

__Polar wind:__ Protons flow upwards on near-vertical magnetic field lines at the poles and are eventually swept away by the solar wind.

### Hydrodynamic escape

Complicated and not well understood: roughly, bulk "blowoff" at the solar wind-atmosphere interface where light atoms moving at sonic speeds drag heavier atoms along with them.

### Impacts

The shock wave from a large meteor strike can cause bulk outflow of a localized column of atmosphere, as well as increased thermal loss over a wider regioun.

### Surface processes

Depending on chemistry, atmospheric components can (at least on rocky planets) be lost downwards by adsorption or dissolution, as well as out to space. An important example is CO<sub>2</sub> dissolving in Earth's oceans.

### International Standard Atmosphere (ISA)

How does the atmosphere vary by height? Saying "it's complicated" is true, but unhelpful for anyone responsible for designing or flying aircraft. The [ISA](https://en.wikipedia.org/wiki/International_Standard_Atmosphere) is defined by [ISO](https://en.wikipedia.org/wiki/International_Organization_for_Standardization) as an idealized, static model which gives a reasonable approximation for clean, dry air across a wide range of latitudes and elevations.

```python
from astropy.constants import g0 # 9.81 m/s
from astropy.table import QTable

from stdatm import StdAtm
```

```python
atm = StdAtm()
heights = (np.linspace(0, 90000, 100))*u.m
data = atm.getResultsQtable(heights)
```

```python
data[::10]
```

#### ISA Pressure and density

These vary more or less smoothly, despite a complicated and discontinuous calculation. On a semi-log plot we can see minor variations from an exponential atmosphere, caused by substantial temperature differences.

```python
fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(111)
ax2 = ax1.twiny()

p_bar = data['pressure'].to(u.bar)
ax1.plot(p_bar, data['height'], 'b-', label='pressure')
levels = atm.height_boundaries.value
P_range = (p_bar.min().value, p_bar.max().value)*u.Pa
ax1.plot(P_range, (levels[1],levels[1])*u.m, 'r:')
ax1.set_xlabel('Pressure (bar)', c='b', fontsize=18)
ax1.set_ylabel('height (m)', fontsize=18)
ax1.xaxis.label.set_color('b')
ax1.tick_params(axis='x', colors='b')
ax1.legend(loc=9)

ax1.text(1e-5, 5000, 'troposphere')
ax1.text(0.01, 11500, 'tropopause', c='red')
ax1.plot(P_range, (levels[4],levels[4])*u.m, 'r:')
ax1.text(1e-5, 25000, 'stratosphere')
ax1.text(0.1, 47500, 'stratopause', c='red')
ax1.plot(P_range, (levels[7],levels[7])*u.m, 'r:')
ax1.text(5e-6, 60000, 'mesosphere')
ax1.text(0.1, 81500, 'mesopause', c='red')
ax1.set_xscale('log')

ax2.plot(data['rho'], data['height'], 'g-.', label='density')
ax2.set_xlabel('density ($kg/m^3$)', c='g', fontsize=18)
ax2.tick_params(axis='x', colors='g')
ax2.set_xscale('log')

plt.title('Pressure and density vs height', fontsize=24, fontweight='bold', y=1.1);
ax2.legend(loc=1);
```

#### ISA Temperature

This is _complicated_, as shown in the plot below. Note that the thermosphere is not included, but the lapse rate is strongly positive (I'm still looking for accurate values; it's not part of the ISA spec) and temperatures can be very high.

```python
fig = plt.figure(figsize=(10, 12))
plt.plot(data['temp_C'], data['height'], 'b-')
levels = atm.height_boundaries.value
T_range = (data['temp_C'].min().value, data['temp_C'].max().value)*u.deg_C
plt.plot(T_range, (levels[1],levels[1])*u.m, 'r:')
plt.text(-80, 5000, 'troposphere')
plt.text(0, 11500, 'tropopause', c='red')
plt.plot(T_range, (levels[4],levels[4])*u.m, 'r:')
plt.text(-80, 25000, 'stratosphere')
plt.text(0, 47500, 'stratopause', c='red')
plt.plot(T_range, (levels[7],levels[7])*u.m, 'r:')
plt.text(-80, 60000, 'mesosphere')
plt.text(0, 85500, 'mesopause', c='red')
plt.xlabel('Temp (C)', fontsize=18)
plt.ylabel('height (m)', fontsize=18);
```
